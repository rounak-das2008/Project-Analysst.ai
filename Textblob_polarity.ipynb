```python
import pandas as pd
```

```python
df = pd.read_csv('Reviews_data.csv')
df.head()
```

```python
from textblob import TextBlob
```

```python
sentiments = []
count = 0
for review in df['Review Body']:
    blob = TextBlob(review)
    polarity = blob.sentiment.polarity

    # Classifying the sentiment
    if polarity > 0:
        sentiment = 'Positive'
    elif polarity < 0:
        sentiment = 'Negative'
    else:
        sentiment = 'Neutral'
    
    sentiments.append(sentiment)
    
    if(count==0):
        print("Started ...")
    elif(count%100 == 0):
        print("~~ 100 sentiments done ..")
    count += 1
```

```python
df['Sentiments'] = sentiments
```
